## k-means 클러스터링

```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snsv
```

```python
#데이터 로드
path_train='C:/Users/hyelim/Desktop/가상데이터2.csv'
df = pd.read_csv(path_train,header=0)
```

```python
df.head()
```

```python
kakao = df['kakao'].sample(frac=1).reset_index(drop=True)
pay= df['pay'].sample(frac=1).reset_index(drop=True)
transport = df['transport'].sample(frac=1).reset_index(drop=True)
taxi= df['taxi'].sample(frac=1).reset_index(drop=True)
star= df['star'].sample(frac=1).reset_index(drop=True)
bean= df['bean'].sample(frac=1).reset_index(drop=True)
ediya = df['ediya'].sample(frac=1).reset_index(drop=True)
bene= df['bene'].sample(frac=1).reset_index(drop=True)
angle= df['angle'].sample(frac=1).reset_index(drop=True)
twosome = df['twosome'].sample(frac=1).reset_index(drop=True)
카페기타= df['카페기타'].sample(frac=1).reset_index(drop=True)
cu = df['cu'].sample(frac=1).reset_index(drop=True)
gs25 = df['gs25'].sample(frac=1).reset_index(drop=True)
seven = df['seven'].sample(frac=1).reset_index(drop=True)
편의점기타 = df['편의점기타'].sample(frac=1).reset_index(drop=True)
```

```python
df = pd.concat([kakao,pay,transport,taxi,star,bean,ediya,bene,angle,twosome,카페기타,cu,gs25,seven,편의점기타], ignore_index=True, axis=1)
```

```python
df
```

```python
df.columns = ['kakao','pay','transport','taxi','star','bean','ediya','bene','angle','twosome','카페기타','cu','gs25','seven','편의점기타']
```

```python
df
```

```python
df.to_csv('C:/Users/hyelim/Desktop/셔플링한 데이터2.csv', index = False)
```
